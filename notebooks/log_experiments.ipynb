```python
log_to_debug = logging.getLogger("log.ui")
while log_to_debug is not None:
    print(f"level: {log_to_debug.level}, name: {log_to_debug.name}, handlers: {log_to_debug.handlers}") 
    log_to_debug = log_to_debug.parent
```

```python
import logging
import yaml

with open('./config/log.yml', 'r') as f:
    config = yaml.safe_load(f.read())
    logging.config.dictConfig(config)

# initiate a new logger
logger = logging.getLogger(__name__)

logger.debug('This is a debug message from config')
logger.info('This is a info log from config')
```

```python
log_file_path = './logs/log.log'
fh = logging.FileHandler(filename=log_file_path)
fh.setFormatter(logging.Formatter(config['formatters']['extended']['format']))
logger.addHandler(fh)

logger.info('This one is from outside of config filehandler')
```

```python
# Changing logging levels
logger.setLevel(logging.INFO)
logger.info('Logger is set to INFO')
logger.debug('This debug message will not be in console but in file')
```
